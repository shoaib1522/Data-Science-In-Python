```python
# Task 8: Select rows with Experience > 5 and Performance Score >= 4
df.loc[(df['Experience'] > 5) & (df['Performance Score'] >= 4)]
```

```python
# Task 10: Select "Sales" employees with Salary in the top 10%
salary_threshold = df['Salary'].quantile(0.9)
df.loc[(df['Department'] == 'Sales') & (df['Salary'] > salary_threshold)]
```

```python
# Task 11: Replace "Inactive" with "Retired"
df['Status'] = df['Status'].replace('Inactive', 'Retired')
```

```python
# Task 12: Add Bonus column as 10% of Salary
df['Bonus'] = df['Salary'] * 0.10
```

```python
# Task 13: Update NaN Performance Scores in "IT" department to 5
df.loc[(df['Department'] == 'IT') & (df['Performance Score'].isna()), 'Performance Score'] = 5
```

```python
# Task 14: Update Experience to 2 for employees with less than 2 years
df.loc[df['Experience'] < 2, 'Experience'] = 2
```

```python
# Task 15: Normalize Salary column
df['Salary_Normalized'] = (df['Salary'] - df['Salary'].min()) / (df['Salary'].max() - df['Salary'].min())
```

```python
# Task 16: Update Location to "Remote" for Performance Score < 3
df.loc[df['Performance Score'] < 3, 'Location'] = 'Remote'
```

```python
# Task 17: Add Experience Level column
df['Experience Level'] = pd.cut(df['Experience'], bins=[0, 3, 7, float('inf')], labels=['Junior', 'Mid', 'Senior'])
```

```python
# Task 18: Add a new row using pd.concat
new_row = pd.DataFrame([{
    'ID': 999, 'Name': 'New Employee', 'Age': 30, 'Gender': 'Other',
    'Department': 'HR', 'Salary': 5000, 'Joining Date': '2024-01-01',
    'Performance Score': 4, 'Experience': 2, 'Status': 'Active',
    'Location': 'New York', 'Session': 'Morning', 'Bonus': 500
}])

df = pd.concat([df, new_row], ignore_index=True)
```

```python
# Task 19: Drop the Session column
df.drop(columns=['Session'], inplace=True)
```

```python
# Task 20: Remove rows where Age > 60
df = df[df['Age'] <= 60]
```

```python
# Task 21: Add Age Group column
df['Age Group'] = pd.cut(df['Age'], bins=[0, 30, 50, float('inf')], labels=['Young', 'Middle-aged', 'Senior'])
```

```python
# Task 22: Find columns with missing values
missing_values = df.isna().sum()
missing_values[missing_values > 0]
```

```python
# Task 23: Fill missing Performance Scores with the column mean
df['Performance Score'] = df['Performance Score'].fillna(df['Performance Score'].mean())
```

```python
# Task 24: Drop rows with missing values
df.dropna(inplace=True)
```

```python
# Task 25: Replace missing Performance Score with median by Department
df['Performance Score'] = df.groupby('Department')['Performance Score'].transform(lambda x: x.fillna(x.median()))
```

```python
# Task 26: Select rows with "IT" employees and Age > average Age in "IT"
average_age_it = df[df['Department'] == 'IT']['Age'].mean()
df.loc[(df['Department'] == 'IT') & (df['Age'] > average_age_it)]
```

```python
# Task 27: Remove duplicates based on Name and Joining Date
df = df.drop_duplicates(subset=['Name', 'Joining Date'])
```

```python
# Task 28: Remove rows with inconsistent Experience
df = df[df['Experience'] <= (df['Age'] - 18)]
```

```python
# Task 29: Sort by Salary in descending order
df = df.sort_values(by='Salary', ascending=False)
```

```python
# Task 30: Sort by Department and Joining Date
df = df.sort_values(by=['Department', 'Joining Date'])
```

```python
# Task 31: Rank employees by Salary within their Department
df['Salary Rank'] = df.groupby('Department')['Salary'].rank(ascending=False)
```

```python
# Task 32: Sort by Experience (descending) and Performance Score (ascending)
df = df.sort_values(by=['Experience', 'Performance Score'], ascending=[False, True])
```

```python
# Task 33: Filter HR employees with Salary > 5000
hr_high_salary = df[(df['Department'] == 'HR') & (df['Salary'] > 5000)]
```

```python
# Task 34: Remove rows where Location is "Chicago"
df = df[df['Location'] != 'Chicago']
```

```python
# Task 35: Remove rows with Performance Score < 2
df = df[df['Performance Score'] >= 2]
```

```python
# Task 36: Calculate average Salary per Department
average_salary_department = df.groupby('Department')['Salary'].mean()
```

```python
# Task 37: Calculate total Salary by Status
total_salary_status = df.groupby('Status')['Salary'].sum()
```

```python
# Task 38: Maximum Performance Score by Location
max_performance_location = df.groupby('Location')['Performance Score'].max()
```

```python
# Task 39: Total Salary by Joining Year
df['Joining Year'] = pd.to_datetime(df['Joining Date']).dt.year
total_salary_year = df.groupby('Joining Year')['Salary'].sum()
```

```python
# Task 40: Department with highest average Performance Score
highest_avg_performance = df.groupby('Department')['Performance Score'].mean().idxmax()
```

```python
# Restore the Session column with default values
df['Session'] = 'Morning'  # You can use actual data or default values
# Task 41: Employee count by Location and Session
employee_count_location_session = df.groupby(['Location', 'Session']).size()
```

```python
# Task 42: Salary stats by Status
salary_stats_status = df.groupby('Status')['Salary'].agg(['sum', 'mean', 'std'])
```

```python
# Task 43: Add Monthly Salary column
df['Monthly Salary'] = df['Salary'].apply(lambda x: x / 12)
```

```python
# Task 44: Convert string values in each column to uppercase
for col in df.select_dtypes(include='object'):  # Select columns with object (string) data type
    df[col] = df[col].map(lambda x: x.upper() if isinstance(x, str) else x)
```

```python
# Task 45: Inner join with Department and Budget dataset
budget_df = pd.DataFrame({'Department': ['Sales', 'HR', 'IT'], 'Budget': [100000, 80000, 120000]})
df = pd.merge(df, budget_df, on='Department', how='inner')
```

```python
# Task 46: Merge with Projects dataset
projects_df = pd.DataFrame({'ID': [1, 2, 3], 'Recent Projects': ['A', 'B', 'C']})
df = pd.merge(df, projects_df, on='ID', how='left')
```

```python
# Task 47: Left join with Team Lead dataset
team_lead_df = pd.DataFrame({'Department': ['Sales', 'HR', 'IT'], 'Team Lead': ['John', 'Sarah', 'Mike']})
df = pd.merge(df, team_lead_df, on='Department', how='left')
```

```python
# Task 48: Merge with Project Count and calculate average Salary per project
project_count_df = pd.DataFrame({'ID': [1, 2, 3], 'Project Count': [3, 5, 2]})
df = pd.merge(df, project_count_df, on='ID', how='left')
df['Salary Per Project'] = df['Salary'] / df['Project Count']
```

```python
# Task 49: Calculate percentage increase in Salary
avg_salary_department = df.groupby('Department')['Salary'].transform('mean')
df['Salary Increase %'] = ((df['Salary'] - avg_salary_department) / avg_salary_department) * 100
```
